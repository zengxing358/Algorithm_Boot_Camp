## 关于对算法的重认知

我以前学习算法主要是由于对逻辑思维的喜爱所引发的兴趣<br>
而在经历了在老师一套方法论引领下的训练之后，逐渐领悟到了一些分析和解决问题的方法，这个在写其他python数据分析代码是，比以往更容易想到一些【偷懒】的方法，这可能是算法的思想或者是算法训练带来的脑力的提升吧<br>
由此，我重新审视了一下【有用的知识】和【无用的知识】的关系，比如哲学就是非常典型的【无用的知识】，其实算法跟他很像，但还是比他有用一些<br>
越是【无用的知识】越是通用性的，因为要适应所有学科，所以显得有些虚，但是掌握了他之后对你去快速了解其他领域助益很大，当让他的功效不能马上显露出来，可以把他认为是一种隐形的帮助，所以也比较容易被忽视，尤其是功利性强的人<br>
越是【有用的知识】越是能够帮助你去解决一个具体的问题，但也是因为具体，他带有很多细节，也就丧失了通用型，他也仅仅只适合某几个场景，他在现在看似很有用，但随着技术的发展，他也很可能将会更新，也就变得无用了<br>
但他变成无用是需要一些时间的，尤其再你遇到问题时他又像救命稻草那样有用，故而让人也比较容易重视他，而且他有足够多的细节，非常便于理解，能够快速上手，这样就让人特别容易忽视了那些【无用的知识】；<br>
尤其是那些【无用的知识】，他需要具有普适性，往往极度抽象，了解他需要费一些脑力，需要经过一系列刻意练习才能逐渐上手，乐意去啃他，确实需要一定兴趣，或者你要解决的问题本身需要应对多个场景，比如编写某个框架；然而在你掌握了他之后，会感觉似乎获得了一种触类旁通的神级力量，他能辅助你以不变应万变<br>
如果说哲学是所有学科中认识世界的一种【无用的知识】，那么算法就是计算机学科中的一种【无用的知识】，也向在我们认知世界的道路上，要不断去拨开重重迷雾，解开真理的面纱一样，想要在计算机科学条路上走的更远，一定要伴着算法一路走下去。算法训练营的结束还只是一个开端，是我在接下来的时间中需要不断打磨的一项技能；<br>
经过这次训练我也找到了自己的不足，也就开始有了一条继续训练的路径，向着刷穿leetcode,刷完Acwing进击，fighting
