# 毕业收获感悟总结

参加训练营之前对于动态规划问题总是不知道思路，课程将动态规划进行了分类之后，感觉好多了，但还是需要不断练习；<br>
对于二分问题之前把三种方法混淆了，每次提交结果都要调试半天，最后也不知道怎么就通过了，这次把这个彻底吃透了；<br>
对于状态和搜索的理解更加深刻了，比如在高级搜索那部分迭代加深，以前在中国大学mooc电子科技大的老师也讲过，但老师说得很含糊把他说成是深度和广度搜索的结合，理解得不是很好，最后相应的题目也是用的A*算法和倍增思想搞定的，而没有再去深究该方法，现在终于明白这个的含义及应用场景；<br>
对于平衡树和线段树总是感到有些害怕，这次训练营分别用python和Java动手实现了，感觉心里有底了很多；<br>
跳表和树状数组是以前未接触过的高级数据结构，这次训练营也动手实现了，感觉记忆深刻，但后续还需要多多练习；<br>
以前感觉leetcode的测试用例很变态，动不动就很长很长的数组，但是知道老师算法进阶题的平台Acwing之后，感觉leetcode的时间限定已经很友好了，此外，以前主流使用语言为python,不需要关注整数是长整型还是短整型，在Acwing很多时候溢出了，找问题是调试大半天，还是多多实操一门强类型语言。<br>
总之，老师的课程不单将数据结构与算法的脉络整体梳理了一遍，让人在宏观上有一个全局概览；还在很多细节处从本质到原理透彻地讲解了很多巧妙而深刻的思想，而这些微观的技巧与思想又贯穿算法与数据结果体系的始末，便于学习和理解新的甚至是更加高级的数据结构与算法，可以作为大体系下的有力支撑。<br>
在参加训练营之前，感觉自己在算法上面似乎已经到达了一个瓶颈，而这次的目标是往300h的降维打击算法面试的。这期间也花了大量时间刷题，代码能力感觉有一定提升，更大的收获是算法的这条路更清晰了，很多问题开始有了一套分析方法。<br>
在后续的训练中，位运算、状态压缩、倍增优化、剪枝还需要刻意练习。另外还需要增加综合训练，不单是把不同类型的题混杂在一起，还包括综合运用各种类型的算法或技巧解题的方法和思想。
